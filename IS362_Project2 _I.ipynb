## Salary Survey

The following analysis was conducted using the salary survey results containing more than 32,000 responses.


```python
import pandas as pd
import numpy as np

salarysurvey = pd.read_csv('salary_survey2019.csv', engine='python')
#Salary of different currency cannot be compared so we will focus on US salary by filtering only USD currency
USsalary = salarysurvey[(salarysurvey["Currency"]=="USD")]
USsalary.head(5)
```

```python
USsalary['Salary'].mean()
```

The average salary of the survey respondants was $103,872. 

This result is highly unrealistic because according to the article below from The Street, the average US income in 2019 was $48, 672.
This means that either most of the survey takers earn above average income or they were not truthful on their responses.

https://www.thestreet.com/personal-finance/average-income-in-us-14852178

## Top 10 highest salary in the US according to this survey

```python
HighestSalary = USsalary.sort_values('Salary', ascending=False)
df = pd.DataFrame(HighestSalary)
output = (df[['Industry', 'Jobtitle', 'Salary']])
output.head(10)
```

## Top 10 lowest salary in the US according to this survey

```python
LowestSalary = USsalary.sort_values('Salary', ascending=True)
df = pd.DataFrame(LowestSalary)
output = (df[['Industry', 'Jobtitle', 'Salary']])
output.head(10)
```

The ten lowest salaries are way to small, leading to the conclusion that some of the survey takers are part-time workers or they confused the annual salary with monthly or weekly pay.
